```python
img_size = 50

ben_training_data_folder = "train/benign/"
mal_training_data_folder = "train/malignant/"

ben_testing_data_folder = "test/benign/"
mal_testing_data_folder = "test/malignant/"

def load_images(folder, label, img_size=50):
    X = []
    y = []
    skipped = 0

    for filename in os.listdir(folder):
        full_path = os.path.join(folder, filename)

        img = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            skipped += 1
            continue
        
        try:
            img = cv2.resize(img, (img_size, img_size), interpolation=cv2.INTER_AREA)
        except cv2.error:
            skipped += 1
            continue

        # 4) Normalize to [0, 1] float32
        img = img.astype(np.float32) / 255.0
        

        X.append(img)
        y.append(label)

    X = np.stack(X, axis=0)               # (N, img_size, img_size, 1)
    y = np.asarray(y, dtype=np.int64)     # (N, 2)

    print(f"{folder}: {len(X)} loaded, {skipped} skipped")
    return X, y

```

```python
# One-hot labels: benign=[1,0], malignant=[0,1]
X_ben_train, y_ben_train = load_images(ben_training_data_folder, [1, 0], img_size)
X_mal_train, y_mal_train = load_images(mal_training_data_folder, [0, 1], img_size)

X_train = np.concatenate([X_ben_train, X_mal_train], axis=0)
y_train = np.concatenate([y_ben_train, y_mal_train], axis=0)

perm = np.random.permutation(len(X_train))
X_train = X_train[perm]
y_train = y_train[perm]

np.save("melanoma_X_train.npy", X_train)
np.save("melanoma_y_train.npy", y_train)
# #do i need to balance the data?
```

```python
# Test data
X_ben_test, y_ben_test = load_images(ben_testing_data_folder, [1, 0], img_size)
X_mal_test, y_mal_test = load_images(mal_testing_data_folder, [0, 1], img_size)

X_test = np.concatenate([X_ben_test, X_mal_test], axis=0)
y_test = np.concatenate([y_ben_test, y_mal_test], axis=0)

perm = np.random.permutation(len(X_test))
X_test = X_test[perm]
y_test = y_test[perm]

np.save("melanoma_X_test.npy", X_test)
np.save("melanoma_y_test.npy", y_test)
```

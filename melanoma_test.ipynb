```python
with torch.no_grad():
    for i in range(len(X_test)):
        logits = net(X_test[i].view(1, 1, img_size, img_size))  # (1,2)
        pred = torch.argmax(logits, dim=1).item()               # 0 or 1
        true = torch.argmax(y_test[i]).item()                   # 0 or 1 (one-hot -> index)

        correct += (pred == true)
        total += 1

print(f"Accuracy: {correct}/{total} = {correct/total:.2%}")
```

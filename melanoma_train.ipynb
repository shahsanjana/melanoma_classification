```python
import torch 
import torch.nn as nn 
import torch.optim as optim
import numpy as np
from nnModel import Net
```

```python
X_train_data = np.load("melanoma_X_train.npy", allow_pickle=True)
y_train_data = np.load("melanoma_y_train.npy", allow_pickle=True)
```

```python
X_train = torch.Tensor([img for img in X_train_data])
y_train = torch.Tensor([dx for dx in y_train_data])
```

```python
net = Net()
optimizer = optim.Adam(net.parameters(), lr=0.001)
loss_function = nn.MSELoss()
```

```python
img_size = 50
batch_size = 100 #images passing through at once 
epochs = 2  #number of times entire dataset is passed through the model
```

```python
for epoch in range(epochs):
    for i in range(0,len(X_train),batch_size):
        print(f"Epoch {epoch+1}, frac {i}/{len(X_train)}")
        batch_X = X_train[i:i+batch_size].view(-1,1,img_size,img_size)
        batch_y = y_train[i:i+batch_size]
        
        #reset gradients to zero before backpropagation
        optimizer.zero_grad()
        outputs = net(batch_X)
        #calculate loss between predicted outputs and actual labels
        loss = loss_function(outputs, batch_y)
        #calculate gradients and update weights
        loss.backward()
        #update the weights of the model based on the calculated gradients
        optimizer.step()
```

```python
torch.save(net.state_dict(), "melanoma_cnn_model.pth")
```
